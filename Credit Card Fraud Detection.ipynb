```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from pandas.api.types import is_sparse
```

```python
#seperating the data for analysis

legit=  creditcard_data[ creditcard_data.Class ==0]
fraud =  creditcard_data[ creditcard_data.Class ==1]
```

### Now we will build a sample dataset where we get similar distribution of legit and fraud transaction.
fraud transaction---->492

```python
legit_sample = legit.sample(n=492)
```

### concatentaing two dataset


```python
new_ds = pd.concat([legit_sample,fraud], axis=0)
```

```python
new_ds.head()
```

```python
new_ds.tail()
```

```python
new_ds['Class'].value_counts()
```

```python
new_ds.groupby('Class').mean()

#as we see the nature of dataset is not changed
```

### Splitting the data into features and targets


```python
X = new_ds.drop(columns='Class' , axis=1)
Y = new_ds['Class']
```

```python
print(X)
```

```python
print(Y)
```

### Split the data into train and test data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

### Model training
mainly we use logistic regression model for binary classifiction model


```python
model = LogisticRegression()
```

```python
# train the data

model.fit(X_train,Y_train)
```

### Model Evalution 
Accuracy score



```python
# accuracy on traing data

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction , Y_train)
```

```python
print('Accuracy on Training data : ',training_data_accuracy )
```

```python
   #accuracy on test data
    
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test Data: ',test_data_accuracy)
```
